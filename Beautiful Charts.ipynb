## Beautiful Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
np.random.seed(0)
y = np.random.rand(5)+3
y
```

```python
plt.bar(x, y)
plt.title('Vertical Bar Chart')
```

```python
# we can combine charts into a figure
plt.subplots(2,2) # a grid of two rows and two columns
# top left
plt.subplot(2,2,1)
plt.pie([6,5,4,3])
# target top right
plt.subplot(2,2,2)
plt.barh(x,y)
# target the bottom left space in the figure
plt.subplot(2,2,3) 
plt.bar(x,y)
# this plots a line in bottom right (2,2,2)
plt.subplot(2,2,4)
plt.plot(x,y) 
```
